## 1.K-Means Clustering

### Finding the optimal number of clusters using Elbow method

```python
# find the within-cluster sum of squared errors for different numbers of clusters
wcss = []

for i in range(1, 11):
    # Create a KMeans model for different clusters
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_)

# Plot wcss against the number of clusters
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster sum of squared errors')

plt.show()
```

### Different parameter configurations



```python
#Silhoutte score of K Means for different PCA components
for x in range(1,14):
    pca_x = PCA(n_components=x)
    pca_x.fit(data_scaled)
    data_pcax = pca_x.transform(data_scaled)

    # Perform K-means clustering on PCA-transformed data
    kmeansx = KMeans(n_clusters=3, random_state=0)
    labels_pcax = kmeansx.fit_predict(data_pcax)

    sil_score = silhouette_score(data_pcax, labels_pcax)
    print("Kmeans for number of pca components ",x," the sil_score: ",sil_score)
```

```python
#Silhoutte score of K Means for different clusters using init='k-means++'
for c in range(2,6):

    # Perform K-means clustering on PCA-transformed data
    kmeans = KMeans(n_clusters=c,init='k-means++', random_state=0)
    labels_pcac = kmeans.fit_predict(data_pca)

    sil_score = silhouette_score(data_pca, labels_pcac)
    print("Kmeans using K-means++ for number of clusters ",c," the sil_score: ",sil_score)
```

```python
#Silhoutte score of K Means for different clusters using init='random'
for c in range(2,6):
    # Perform K-means clustering on PCA-transformed data
    kmeans = KMeans(n_clusters=c,init='random', random_state=0)
    labels_pcac = kmeans.fit_predict(data_pca)

    sil_score = silhouette_score(data_pca, labels_pcac)
    print("Kmeans using random for number of clusters",c," the sil_score: ",sil_score)
```

### Training the model

```python

# Perform K-means clustering on PCA-transformed data
kmeans = KMeans(n_clusters=3,init='k-means++', random_state=0)
labels_Kpca = kmeans.fit_predict(data_pca)

# Perform K-means clustering on original data
kmeans = KMeans(n_clusters=3, random_state=0)
labels_Koriginal = kmeans.fit_predict(data_scaled)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the results of K-means clustering on original data
ax1.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels_Koriginal)
ax1.set_title('K-means clustering without PCA')

# Plot the results of K-means clustering on PCA-transformed data
ax2.scatter(data_pca[:, 0], data_pca[:, 1], c=labels_Kpca)
ax2.set_title('K-means clustering with PCA')

plt.show()
```

## 2.Gaussian Mixture

### Finding the optimal number of clusters using GMM ( Having low AIC and BIC values)

```python
# approach followed from https://jakevdp.github.io/PythonDataScienceHandbook/05.12-gaussian-mixtures.html
n_clusters = np.arange(1, 11)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(data_pca[:,:2])
          for n in n_clusters]

plt.plot(n_clusters, [m.bic(data_pca[:,:2]) for m in models], label='BIC')
plt.plot(n_clusters, [m.aic(data_pca[:,:2]) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_clusters');
```

```python
# approach followed from https://jakevdp.github.io/PythonDataScienceHandbook/05.12-gaussian-mixtures.html
# GMM Plot having full covariance type
def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()

    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)

    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))

def plot_gmm(gmm, X, label=True, ax=None):
    ax = ax or plt.gca()
    labels = gmm.fit(X).predict(X)
    if label:
        ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
    else:
        ax.scatter(X[:, 0], X[:, 1], s=40, zorder=2)
    ax.axis('equal')
    ax.set_title('GMM Clustering with full variance')
    w_factor = 0.2 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        draw_ellipse(pos, covar, alpha=w * w_factor)
```

```python
pca1 = PCA(n_components=0.99)
pca1.fit(df_scaled1)
data_pca1 = pca1.transform(df_scaled1)
```

### Finding the exact eps and minimum samples

Before passing our parameters to Dbcsan model we have to find the right eps and minimum samples.So, below we have taken the help of Nearest Neighbours to get our eps values.


```python
#approach followed from https://towardsdatascience.com/machine-learning-clustering-dbscan-determine-the-optimal-value-for-epsilon-eps-python-example-3100091cfbc
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt
import numpy as np
neighbors = NearestNeighbors(n_neighbors=100)
neighbors_fit = neighbors.fit(data_pca1)
distances, indices = neighbors.kneighbors(data_pca1)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

As, we can see from above plot the cuve is bending near 2.3 - 2.4. So, this is the eps value which we used in our below for loop to get the exact min samples.

```python
# Checking for the right eps
for min_samples in range(3,40):
    print("\nmin_samples={}".format(min_samples))
    dbscan = DBSCAN(eps=2.35, min_samples=min_samples)
    labels1 = dbscan.fit_predict(data_pca1)
    print("Number of clusters: {}".format(len(np.unique(labels1))))
    print("Cluster sizes: {}".format(np.bincount(labels1 + 1)))
```

Minimum samples as estimated above we got range between the 12 - 18 values and the then final parameters has been passed to below parameters.

### Training the Model

```python

# Fit the DBSCAN model
dbscan = DBSCAN(eps=2.35, min_samples=12)
dbscan.fit(data_pca1)
model = dbscan.fit_predict(data_pca1)

# Extract the clusters and the noise points
labels1 = dbscan.labels_
print("Clusters: ", labels1)
print("Total Number of cluster's created :" +str(list(set(model))))
```

The plot below gives us exact of 3 clusters so to get clear view of the clusters we have excluded the noise points(-1) from it.

```python
# Plot the wine dataset using the DBSCAN labels, excluding noise points
fig, ax = plt.subplots(figsize=(10, 8))
scatter = ax.scatter(data_pca1[dbscan.labels_ != -1, 0], data_pca1[dbscan.labels_ != -1, 1],
                     c=dbscan.labels_[dbscan.labels_ != -1], cmap='viridis', label="Cluster")
plt.title("Dbscan")
plt.show()
```

## 6.Ada Boosting Classifier

AdaBoost is a boosting technique that combines several weak learners to create a strong learner. It is a sequential strategy that works by first fitting a classifier on the original dataset and then fitting multiple copies of the classifier on the same dataset, but with the weights of erroneously classified instances changed such that succeeding classifiers focus on difficult situations.

### Split data protocol

The X and Y data must then be used to generate training and testing sets. By setting the test size argument to 0.33, we'll use 33% of the data for testing. We'll also set the random state argument to 1 to ensure that the split is the same every time we execute the function.

```python
X6=wine.data
scaler = StandardScaler()
X6 = scaler.fit_transform(X6)
y6=wine.target
X6_train,X6_test,y6_train,y6_test = train_test_split(X6,y6,test_size=0.33,random_state=0)
```

### Training

We can now create our AdaBoost classification model using the AdaBoostClassifier class. To compare accuracy, we'll set the n estimators option to 100 and compare the base estimator RandomForest with the default Descion Tree. The model will then be trained using the fit strategy on the training data.

```python
#Create adaboost classifier object
from sklearn.ensemble import AdaBoostClassifier
AdaModel=AdaBoostClassifier(n_estimators=100,random_state=0)
AdaModel1=AdaBoostClassifier(n_estimators=100,random_state=0,base_estimator=RandomForestClassifier())
```

```python
#Train Adaboost Classifier
model=AdaModel.fit(X6_train,y6_train)
model1=AdaModel1.fit(X6_train,y6_train)
#Predict the response of test dataset

```

By sending the X test data to the predict function, we can now make predictions on the test data. The predictions will be assigned to the y pred variable. They are returned in the form of a NumPy array.

```python
y6_pred=model.predict(X6_test)
y6_pred1=model1.predict(X6_test)
y6_pred1
```

### Evaluation

To evaluate the model, we may utilise the accuracy score function. This function compares the expected and actual numbers and returns the accuracy score as a percentage. To gain a more complete analysis of the model's performance, we'll also utilise the classification report function. The early results aren't really impressive, but the Random option produces the best outcomes.

```python
#Model Accuracy
print("Accuracy with Default Decsion Tree Model :",metrics.accuracy_score(y6_test,y6_pred))
print("Accuracy with Random Forest Model :",metrics.accuracy_score(y6_test,y6_pred1))
```
